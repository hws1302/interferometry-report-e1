```python
df = pd.read_csv('velocitydetection2/velocitydetection2_17.csv', skiprows=[1], index_col=0)

df.plot(title='11',)
```

```python
df
```

```python
N = 100
SAMPLE_RATE = 0.040960 # seconds


yf = rfft(list(df['Channel A'])[1000:1100]) # using rfft saves on compute 
Yf = np.abs(yf)
xf = rfftfreq(N, 1/SAMPLE_RATE)

plt.plot(xf, np.abs(yf))
plt.show()

len(xf)
```

```python
max_index = np.where(Yf[1:] == max(Yf[1:]))

max_value = xf[max_index]

# surf logarithmic scale 

max_index
```

```python
x = np.linspace(0,4500,45)
y = []

for i in range(0,4500,100):
    
    yf = rfft(list(df['Channel A'])[i:i+100]) # using rfft saves on compute 
    Yf = np.abs(yf)
    xf = rfftfreq(N, 1/SAMPLE_RATE)
    max_index = np.where(Yf[1:] == max(Yf[1:]))

    max_value = xf[max_index]
    
    y.append(max_value)
    
```

```python
plt.plot(x,y)
```
