```python
# we care about x prefactors but not y prefactors 
from scipy.fft import fft, fftfreq, rfft, rfftfreq, irfft
from scipy.io.wavfile import write
import numpy as np
import matplotlib.pyplot as plt 
from tqdm import tqdm 
```

```python
# endpoint=False to show periodic 
SAMPLE_RATE = 44100 #Hertz
DURATION = 5 #Seconds
FREQ = 2 #Hertz

def generate_sine_wave(freq, sample_rate, duration):
    x = np.linspace(0,duration,sample_rate*duration, endpoint=False)
    y = np.sin(2* np.pi * freq *x) 
    return x, y 

x, y = generate_sine_wave(FREQ, SAMPLE_RATE, DURATION)

plt.plot(x,y)
plt.show()
```

```python
_, nice_tone = generate_sine_wave(400, SAMPLE_RATE, DURATION)
_, noise_tone = generate_sine_wave(4000, SAMPLE_RATE, DURATION)
noise_tone = noise_tone * 0.3

mixed_tone = nice_tone + noise_tone

normalized_tone = np.int16((mixed_tone / mixed_tone.max()) * 32767)

plt.plot(normalized_tone[:1000])
plt.show()
```

```python
write('mysinewave.wav', SAMPLE_RATE, normalized_tone)
```

```python
# now to start with the fft

N = SAMPLE_RATE * DURATION

yf = fft(normalized_tone)
xf = fftfreq(N, 1/SAMPLE_RATE)
```

```python
plt.plot(xf, np.abs(yf))
plt.show()
```

```python
# if you have a real signal then can use rfft to get 
# rid of the mirror symmetry but then it mirrors at end

yf = rfft(normalized_tone)
xf = rfftfreq(N, 1/SAMPLE_RATE)
plt.plot(xf, np.abs(yf))
plt.show()
```

```python
# would be good to use tqdm to see how long 
# fft takes vs. rfft
```

```python
# what is this doing here?

points_per_freq = len(xf) /(SAMPLE_RATE/2)

target_idx = int(points_per_freq * 4000)

yf[target_idx - 1 : target_idx + 2] = 0

plt.plot(xf, np.abs(yf))
plt.show()
```

```python
# Applying the inverse fft 

new_sig = irfft(yf) 
norm_new_sig = np.int16(new_sig * (32767 / new_sig.max()))

write("clean.wav", SAMPLE_RATE, norm_new_sig)
plt.plot(new_sig[:1000])
```
